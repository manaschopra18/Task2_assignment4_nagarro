```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("testset.csv",parse_dates=['datetime_utc'])
```

```python
dataset
```

```python
sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False,cmap='viridis')
#Here yellow color signifies NaN values in data
```

```python
dataset.columns
```

```python
dataset.drop([' _heatindexm',' _precipm',' _wgustm',' _windchillm'],axis=1,inplace=True)
```

```python
sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
dataset.columns
```

```python
dataset[' _dewptm'].fillna(dataset[' _dewptm'].mean(),inplace=True)
dataset[' _hum'].fillna(dataset[' _hum'].mean(),inplace=True)
dataset[' _pressurem'].fillna(dataset[' _pressurem'].mean(),inplace=True)
dataset[' _tempm'].fillna(dataset[' _tempm'].mean(),inplace=True)
dataset[' _vism'].fillna(dataset[' _vism'].mean(),inplace=True)
dataset[' _wdird'].fillna(dataset[' _wdird'].mean(),inplace=True)
dataset[' _wdird'].fillna(dataset[' _wdird'].mean(),inplace=True)
dataset[' _wspdm'].fillna(dataset[' _wspdm'].mean(),inplace=True)
```

```python
dataset[' _wdire'].fillna((dataset[' _wdire'].mode()[0]),inplace=True) #Mode because type=String
```

```python
sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False,cmap='viridis')
#Here yellow color signifies NaN values in data
```

#### Removing Outliers

```python
x=[' _dewptm', ' _fog', ' _hail', ' _hum',' _pressurem', ' _rain', ' _snow', ' _tempm', ' _thunder', ' _tornado',' _vism', ' _wdird',' _wspdm']
for i in x:
    dataset[i] = dataset[i][dataset[i].between(dataset[i].quantile(.025), dataset[i].quantile(.975))]
```

```python
dataset=dataset.dropna()
```

```python
dataset['Month']=dataset['datetime_utc'].apply(lambda x: x.month)
```

```python
dataset.head(2)
```

```python
dataset['Year']=dataset['datetime_utc'].apply(lambda x: x.year)
```

```python
dataset.corr()
```
