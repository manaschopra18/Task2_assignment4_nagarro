### Q1- Which team has won/lost maximum number of matches while chasing?

### FINDING MAX WINNER WHILE CHASING

```python
max_wins=matches[((matches['toss_winner']==matches['winner']) & (matches['toss_decision']=='field'))| ((matches['toss_winner']!=matches['winner'])& (matches['toss_decision']=='bat'))].groupby('winner').count()['id']
```

```python
pd.DataFrame(max_wins).rename(columns={'id':'Count'}).sort_values("Count",ascending=False).iloc[0]
```

### FINDING MAX LOSE WHILE CHASING

```python
losers_condition_1=matches[(matches['toss_winner']!=matches['winner']) & (matches['toss_decision']=='field')].groupby('toss_winner').count()['id']
```

```python
losers_condition_2=matches[(matches['toss_winner']==matches['winner'])& (matches['toss_decision']=='bat')][['team1','team2','winner']]
```

```python
losers_condition_2_1=losers_condition_2[losers_condition_2['team1']==losers_condition_2['winner']].groupby('team2').count()['team1']
```

```python
losers_condition_2_2=losers_condition_2[losers_condition_2['team2']==losers_condition_2['winner']].groupby('team1').count()
```

```python
ans=pd.DataFrame(losers_condition_1+losers_condition_2_1).reset_index()
```

### Q9 - ( iV ) Which umpire has umpired maximum matches?

```python
ump1=matches[matches['umpire1']!=""][['umpire1','umpire2']].groupby('umpire1').count().reset_index().rename(columns={'umpire1':'UMPIRES','umpire2':'count'})
```

```python
ump2=matches[matches['umpire2']!=""][['umpire2','umpire1']].groupby('umpire2').count().reset_index().rename(columns={'umpire2':'UMPIRES','umpire1':'count'})
```

```python
ump3=matches[matches['umpire3']!=""][['umpire3','umpire1']].groupby('umpire3').count().reset_index().rename(columns={'umpire3':'UMPIRES','umpire1':'count'})
```

```python
ump1.append(ump2).append(ump3).groupby('UMPIRES').sum().sort_values(by='count').iloc[-1]
```
